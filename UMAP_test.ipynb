```python
import umap
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('master_df_reduced.csv')
data.columns
```

```python
data_umap = data[[
    'Latitude',
    'Longitude',
    'Focal Depth (km)',
    'Mag',
    'Average Income',
    'p0p50_share',
    'pop_total',
    # 'badness_value',
]]
```

```python
data_map = umap.UMAP(n_components=2, min_dist=0.8).fit_transform(data_umap)
```

```python
ax = sns.scatterplot(x=data_map[:,0], y=data_map[:,1], hue=data.badness_value,palette='viridis', alpha=0.5)
sns.move_legend(ax, "upper left", bbox_to_anchor=(1, 1))
ax.set(title='UMAP dimension reduction')
fig = ax.get_figure()
fig.savefig('Plots/UMAP.pdf', bbox_inches='tight')
```
